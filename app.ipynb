```python
from model.app import FaceSwapper
from model.Model import VAE
import torch
import os
from PIL import Image
import matplotlib.pyplot as plt

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
model = VAE(device).to(device).eval()
model.load_state_dict(torch.load(os.path.join('model', 'vae.pth'), map_location=device.type))
encoder = model.encoder
```

```python
dst_image = Image.open('faces_one/189.jpg')
src_image = Image.open('faces_one/5654.jpg')

obj = FaceSwapper(encoder, device)
result = obj.swap(src_image, dst_image, verbose=False)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 15))

axes[0].imshow(src_image)
axes[0].axis('off')
axes[0].set_title('Source')

axes[1].imshow(dst_image)
axes[1].axis('off')
axes[1].set_title('Destination')

axes[2].imshow(Image.fromarray(result))
axes[2].axis('off')
_ = axes[2].set_title('Swapped Src->Dst')
```
